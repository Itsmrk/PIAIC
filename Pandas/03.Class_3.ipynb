```python
df1 = pd.DataFrame({"CNIC": [1, 3, 4], "A":[1, 3, 23]})
df2 = pd.DataFrame({"CNIC": [1, 2, 3], "B":[1, 3, 23]})
df3 = pd.DataFrame({"CNIC": [1, 4, 7], "C":[1, 3, 23]})

df = df1.merge(df2, on = "CNIC", how = 'left')
df = df.merge(df3, on = "CNIC", how = 'left')
df
```
