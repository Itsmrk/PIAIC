```python
# A.I -> SUM
# IOT -> Product
# CNC -> cumsum
# else -> 0

def custum_fn(x1, *l1):
    if x1 =='A.I':
        return np.sum(l1)
    elif x1 == "IOT":
        return np.product(l1)
    elif x1 == "CNC":
        return sum(np.cumsum(l1))
    else:
        return 0
```
